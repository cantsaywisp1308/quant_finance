```python
from fredapi import Fred
from fred_api_key import API_KEY
```

```python
fred = Fred(api_key=API_KEY)
```

```python
import pandas as pd
sp500 = pd.DataFrame(fred.get_series('SP500'), columns=['SP500'])
sp500
```

```python
ffr = pd.DataFrame(fred.get_series('FEDFUNDS'), columns=['FEDFUNDS'])
ffr
```

```python
vix = pd.DataFrame(fred.get_series('VIXCLS'), columns=['VIX'])
vix
```

```python
df = pd.concat([sp500, vix])
df
```

```python
import matplotlib.pyplot as plt
plt.plot(df['SP500'], label='SP500')
plt.grid(True)
plt.show()
```
